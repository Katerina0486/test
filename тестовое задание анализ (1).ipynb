Просмотр самых продоваемых товаров на сайтах

Сводная таблица для лучшей работы

```python
data_1 = data[['Товар (офф. Наименование)','Сайт']]
data_1.head(50)
```

```python
data_2 = data_1.pivot_table(index = 'Сайт', values = 'Товар (офф. Наименование)', aggfunc = 'count')\
.sort_values(by='Товар (офф. Наименование)',ascending=False)
data_2
```

И так у каждого сайта имеются продажи более 1000 раз

```python
data_2.plot(y='Товар (офф. Наименование)', kind='bar', figsize=(15, 9),\
            ylabel='Количество товаров проданных на сайте ', xlabel='Сайты')
plt.title(label = 'Продажи по сайтам')
plt.show()
```

Часто покупаемые товар на сайтах

```python
data_3 = data_1.pivot_table(index = 'Товар (офф. Наименование)',values = 'Сайт', aggfunc = 'count')\
.sort_values(by='Сайт',ascending=False).head(10)
data_3
```

```python
data_3.plot(y='Сайт', kind='bar', figsize=(15, 9), ylabel='Сайт', xlabel='Популярные товары')
plt.title(label = 'Топ товаров проданных на сайтах')
plt.show()
```
